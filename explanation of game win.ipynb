```python
%pylab inline

import numpy as np
import pandas as pd
import math as m
import cvxpy as cvx
from sklearn.metrics import accuracy_score,confusion_matrix
import random as rnd
import functools as fu

indxes = [str(x) for x in range(1,114)]
heroes = ["Victory","Ping","Radiant","Dire"]+indxes
train = pd.read_csv("d2Train.csv",names=heroes)
test  = pd.read_csv("d2Test.csv",names=heroes)

X_train = train[indxes]
Y_train = train["Victory"]

X_test = test[indxes]
Y_test = test["Victory"]

fs_train = np.matrix(X_train).T
ts_train = np.matrix(Y_train).T
fs_test = np.matrix(X_test).T
ts_test = np.matrix(Y_test).T

d_margin = ts_train.T*fs_train.T;
```

## Interim functions

```python
def margin(w,v,z):
    return w.T*v*z
```

```python
def objective_func(w):
    return m.log(1+m.exp(-(margin(w,fs_train,ts_train))),m.e)

def dobjective_func(w):
    return (-d_margin*(m.exp(-margin(w,fs_train,ts_train)))/(1+m.exp(-margin(w,fs_train,ts_train)))).T
```

```python
def d(X):
    return 0.5*cvx.norm(X)**2

def dd(X):
    return X
```

### Additional distribution of teams

```python
def additional_features(df):
    rad_dums =  pd.get_dummies(df["Radiant"])
    dire_dums = pd.get_dummies(df["Dire"])
    rad_dums.columns =  ["R"+str(i) for i in range(1,10)]
    dire_dums.columns = ["D"+str(i) for i in range(1,4)]
    return pd.concat([rad_dums,dire_dums],axis=1)
```

# Fast Gradient Method

### Model on addiatonal distrubition

```python
X_train = train[indxes]
Y_train = train["Victory"]
X_train = pd.concat([X_train,additional_features(train)],axis=1)


X_test = test[indxes]
Y_test = test["Victory"]
X_test = pd.concat([X_test,additional_features(test)],axis=1)


fs_train = np.matrix(X_train).T
ts_train = np.matrix(Y_train).T
fs_test = np.matrix(X_test).T
ts_test = np.matrix(Y_test).T

d_margin = ts_train.T*fs_train.T;

n = X_train.shape[1]
x = np.matrix([1/n]*n).T
k = 0
L = 1000
a_vs = [1/L]
y_vs = [x]
u_vs = [x]
x_vs = [x]
while True:
    print(k)
    sum_a = sum(a_vs)
    a_vs.append((1+m.sqrt(1+4*L*L*a_vs[k]*a_vs[k]))/(2*L))
    y_vs.append((a_vs[k+1]*u_vs[k]+sum(a_vs)*x_vs[k]))
    
    x = cvx.Variable(n)
    objective = cvx.Minimize((a_vs[k+1]*dobjective_func(y_vs[k+1]) -dd(u_vs[k])).T*x+d(x))
    prob = cvx.Problem(objective)
    prob.solve()
    u_vs.append(x.value)
    
    x_vs.append((a_vs[k+1]*u_vs[k+1]+sum_a*x_vs[k])/sum(a_vs))
    
    if abs(objective_func(x_vs[k+1]) - objective_func(x_vs[k]) < 0.00000000000000000000000000000000000000001):
        break
    k+=1
test_status = sign(x_vs[k+1].T*fs_test).T
print("Confusion matrix is:")
print(confusion_matrix(test_status,ts_test))
print("Accuracy score is:",accuracy_score(test_status,ts_test))

first_method = x_vs[k+1]
```

# Mirrors Descent Method

### None smooth functions

```python
def obj_func(X):
    return 2*cvx.norm(1-margin(X,fs_train,ts_train),2)
def dobj_func(X):
    return d_margin.T*(1/np.sqrt(1-margin(X,fs_train,ts_train))).T*(1/np.sqrt(1-margin(X,fs_train,ts_train)))
```

### Baseline model working only on heroes distribution

```python
X_train = train[indxes]
Y_train = train["Victory"]

X_test = test[indxes]
Y_test = test["Victory"]

fs_train = np.matrix(X_train).T
ts_train = np.matrix(Y_train).T
fs_test = np.matrix(X_test).T
ts_test = np.matrix(Y_test).T 

d_margin = ts_train.T*fs_train.T;


n = X_train.shape[1]
L  = 10000
x = np.matrix([0]*n).T
c_s = [np.random.uniform(low=0.0001, high=0.01, size=(1,))[0]] 
w_s = [x]
k = 0
f_t = 1000000
w_t = x


while True:
    c_s.append(np.random.uniform(low=0.0001, high=0.01, size=(1,))[0])
    t = cvx.Variable(n)
    objective = cvx.Minimize((c_s[k]*dobj_func(w_s[k]) - dd(w_s[k])).T*t+d(t))
    
    prob = cvx.Problem(objective)
    prob.solve()
    w_s.append(t.value)
    #print(k)
    
    agg = list((map((lambda x :x[0]*x[1]),zip(w_s,c_s))))
    x_t = sum(c_s)**(-1)*fu.reduce(lambda a,b: a+b, agg[1:],agg[0])
    
    if (f_t > obj_func(w_s[k]).value):
        w_t = w_s[k]
    f_v = max(obj_func(w_t).value,obj_func(x_t).value)
    
    if ((-f_v + obj_func(w_s[k+1]).value) <((-d(x).value + d(w_s[k+1]).value + dd(x).T*(w_s[k+1]-x) + 0.5*L*L*sum(list(map((lambda x:x**2),c_s))))/sum(c_s))):
        break
    k+=1

test_status = sign(w_s[k+1].T*fs_test).T
print("Confusion matrix is:")
print(confusion_matrix(test_status,ts_test))
print("Accuracy score is:",accuracy_score(test_status,ts_test))
third_method = w_s[k+1]
    
```
